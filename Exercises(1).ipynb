### Step 7. Which was the most-ordered item?

```python
c = chipo.groupby('item_name')
c = c.sum()
c = c.sort_values(['quantity'], ascending=False)
c.head(1)
```

### Step 8. How many items were orderd in total?

```python
total_items = chipo.quantity.sum()
total_items
```

### Step 9. Turn the item price into a float

#### Step 9.a. Check the item price type

```python
chipo.item_price.dtype
```

#### Step 9.b. Create a lambda function and apply it to the item price

```python
dollarizer = lambda x: float(x[1:-1])
chipo.item_price = chipo.item_price.apply(dollarizer)
```

#### Step 9.c. Check the item price type

```python
chipo.item_price.dtype
```

### Step 10. How many (unique) orders were made in total?

```python
orders = chipo.order_id.value_counts().count()
orders
```

### Step 11. What is the average revenue amount per order?

```python
chipo.groupby(by=['order_id']).sum().mean()['revenue']
```
